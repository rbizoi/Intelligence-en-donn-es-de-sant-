## Clé primaire pour la table COMMANDES01

```python
%sql ALTER TABLE COMMANDES01 ADD CONSTRAINT COMMANDES01_pk PRIMARY KEY ( no_commande )
```

```python
%%sql
select     tc.table_schema, 
    tc.constraint_name, 
    tc.table_name, 
    cc.column_name
from information_schema.constraint_column_usage cc
     join information_schema.table_constraints tc
     on cc.table_catalog = tc.table_catalog
        and cc.table_schema    = tc.table_schema
        and cc.table_name      = tc.table_name
        and cc.constraint_name = tc.constraint_name
where tc.table_name = 'commandes01'
```

## Clé étrangère référencent la colonne NO_EMPLOYE de la table EMPLOYES

```python
%sql ALTER TABLE COMMANDES01 ADD CONSTRAINT COMMANDES01_EMPLOYES_fk FOREIGN KEY ( NO_EMPLOYE ) REFERENCES EMPLOYES ( NO_EMPLOYE )
```

```python
%sql ALTER TABLE COMMANDES01 ADD CONSTRAINT COMMANDES01_CLIENTS_fk FOREIGN KEY ( CODE_CLIENT ) REFERENCES CLIENTS ( CODE_CLIENT )
```

```python
%%sql 
SELECT
    tc.table_schema, 
    tc.table_name, 
    kcu.column_name, 
    tc.constraint_name, 
    ccu.table_schema AS foreign_table_schema,
    ccu.table_name AS foreign_table_name,
    ccu.column_name AS foreign_column_name 
FROM 
    information_schema.table_constraints AS tc 
    JOIN information_schema.key_column_usage AS kcu
      ON tc.constraint_name = kcu.constraint_name
      AND tc.table_schema = kcu.table_schema
    JOIN information_schema.constraint_column_usage AS ccu
      ON ccu.constraint_name = tc.constraint_name
      AND ccu.table_schema = tc.table_schema
WHERE tc.constraint_type = 'FOREIGN KEY' 
  and tc.table_name = 'commandes01'
```
